```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
# Data extraction
#import pandas
import pandas as pd
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
# load dataset
pima = pd.read_csv("D:/sshrivastava/Desktop/Noida/Machine Learning/Example-4-Logistic-Regression.csv", header=None, names=col_names)
```

```python
pima.head()
```

```python
#split dataset in features and target variable
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
X = pima[feature_cols] # Features
y = pima.label # Target variable
```

```python
# split X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
# Model Development and Prediction
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression()

# fit the model with data
model=logreg.fit(X_train,y_train)
model.intercept_
```

```python
model.coef_
```

```python
#to get the probability in each class, for example, first column is probability of y=0 and 
# second column is probability of y=1.

model.predict_proba(X_test)
```

```python
prob1=model.predict_proba(X_test)[:,1]
```

```python
prob1
```

```python
predicted=[1 if i > 0.60 else 0 for i in prob1]
```

```python
predicted
```

```python
# import the metrics class
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, predicted)
```

```python
cnf_matrix
```

```python
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
# Confusion Matrix Evaluation Metrics
print("Accuracy:",metrics.accuracy_score(y_test, predicted))
```
